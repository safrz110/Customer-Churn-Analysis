```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("/Users/sarfarazali/Downloads/Customer Churn Analysis/telecom_churn_data.csv")
df.head(10)
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
df.duplicated()
```

```python
# Create churn column
df['churn'] = np.where(df['arpu_9'] == 0, 1, 0)
df['churn'].value_counts()
```

```python
# Drop unnecessary columns 
if 'mobile_number' in df.columns:
    df.drop('mobile_number', axis=1, inplace=True)
```

```python
# Convert date columns
date_cols = [
    'last_date_of_month_6',
    'last_date_of_month_7',
    'last_date_of_month_8',
    'last_date_of_month_9'
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col])
```

```python
# Handle missing values

 # Drop columns with >70% missing values
null_percent = df.isnull().mean() * 100
cols_to_drop = null_percent[null_percent > 70].index
df.drop(columns=cols_to_drop, inplace=True)

# Fill remaining nulls
df.fillna(0, inplace=True)

print("Remaining nulls:", df.isnull().sum().sum())
```

```python
# Feature Engineering

# Average ARPU (June–Aug)
df['avg_arpu'] = (df['arpu_6'] + df['arpu_7'] + df['arpu_8']) / 3

# ARPU drop
df['arpu_drop'] = df['arpu_8'] - df['arpu_9']

# Total data usage
df['total_data'] = (
    df.get('jun_vbc_3g', 0) +
    df.get('jul_vbc_3g', 0) +
    df.get('aug_vbc_3g', 0)
)
```

```python
# Exploratory Data Analysis (EDA)

 # Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
# Average ARPU vs Churn
plt.figure(figsize=(6,4))
sns.boxplot(x='churn', y='avg_arpu', data=df)
plt.title("ARPU vs Churn")
plt.show()
```

```python
# ARPU Drop vs Churn
plt.figure(figsize=(6,4))
sns.boxplot(x='churn', y='arpu_drop', data=df)
plt.title("ARPU Drop vs Churn")
plt.show()
```

```python
# Data Usage vs Churn
plt.figure(figsize=(6,4))
sns.boxplot(x='churn', y='total_data', data=df)
plt.title("Data Usage vs Churn")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
